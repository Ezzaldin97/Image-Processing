```python
!pip install patchify
import numpy as np
import cv2
import matplotlib.pyplot as plt
from patchify import patchify, unpatchify
import os
```

```python
orig_img = cv2.imread('/content/136762_sat.jpg')
orig_img = cv2.cvtColor(orig_img, cv2.COLOR_BGR2RGB)
orig_img.shape
```

```python
plt.figure(figsize=(15, 15))
plt.imshow(orig_img)
plt.show()
```

<p>as we can see the dimension of the image is very large compared to the dimensions that our CNNs accept to we have to split the image into smaller batches before feeding it to the network.</p>

```python
# using patchify to split the images into patches of 256 by 256
patches = patchify(orig_img, (256, 256, 3), step = 64) #Step=256 for 256 patches means no overlap
patches.shape
```

```python
folder_name = '136762_sat'
folder_path = os.path.join('/content', folder_name)
os.mkdir(folder_path)
print(folder_path)
patch_num = 1
for w in range(patches.shape[0]):
  for h in range(patches.shape[1]):
    single_img = patches[w, h, 0, :, :, :]
    splited_img = cv2.imwrite(os.path.join(folder_path, f'img{patch_num}.jpg'), single_img)
    print(os.path.join(folder_path, f'img{patch_num}.jpg'))
    print("Finished processing patch number ", patch_num, " at position ", w,h)
    patch_num+=1
```
